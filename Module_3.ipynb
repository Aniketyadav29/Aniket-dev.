# **Module 3** _Data Frame Information.

```python
import pandas as pd

df = pd.read_csv("/content/student.csv")

df.info() # displays a complete summary of the data frame
print(df.shape) # returns the number of rows and columnns
print(df.columns) # shows all columns
print(df.index) # displays rows laber (index).
print(df.dtypes) # shows the data type of each column
print(df.size) # returns the total number of elements
print(df.ndim) # returns the number of dimensions.
print(df.memory_usage) # displays memory used by each column.
print(df.describe()) # provides statistical summary of numeric columns.
print(df.describe(include='all')) # shows statics for all columns (numeric nd categorical)
print(df.count()) # counts non-mising values in each column
print(df.isnull().sum()) # counts mising values
print(df.nunique()) # shows the number of unique values in each column.
print(df.value_counts()) # counts unique rows.
```

```python
import pandas as pd

df = pd.read_csv("/content/student.csv")

print(df["Gender"]) # Selecting a Single Column
print(df[["Education", "City"]]) # Selecting multiple columns

# Selecting Rows using Index (iloc uses integer positions)
print(df.iloc[0]) # first row
print(df.iloc[0:5]) # multiple rows (indices 0, 1, 2, 3, 4)
print(df.iloc[-5:]) # last five rows

## Selecting Rows and Columns using loc[]
# loc[] uses row labels (index names) and column names.
# syntax -> df.loc[row_label, column_name]
print(df.loc[0]) # first row (assuming integer label 0 exists)
print(df.loc[0:5]) # multiple rows (labels 0 through 5 INCLUSIVE)
# print(df.loc[-5:]) # ERROR: loc doesn't support negative indexing on default integer indices. Use iloc instead.


# Boolean Indexing
# Boolean Indexing returns rows where the condition is True.

print(df[df["Age"] > 30]) # Age greater than 30
print(df[df["Age"] < 30]) # Age less than 30
print(df[df["City"] == "Bangalore"]) # city is Bangalore
print(df[df["Education"] == "Masters"]) # education is Masters

# Filtering Data

print(df[df["City"] == "Bangalore"]) # single condition

# Fixed multiple conditions (AND) - removed stray parenthesis
print(df[(df["City"] == "Pune") & (df["Age"] > 30)])

# Fixed multiple conditions (OR) - corrected parentheses alignment
print(df[(df["City"] == "Pune") | (df["Age"] > 30)])

print(df[~(df["City"] == "New Delhi")]) # NOT condition

print(df[df["City"].isin(["Bangalore", "Pune"])]) # using isin()

print(df[df["Age"].between(25, 35)]) # using between()

# using str.contains() (added na=False to handle missing values safely)
print(df[df["Education"].str.contains("Bachelor", case=False, na=False)])

# using query
print(df.query("Age > 30"))
print(df.query("City == 'Pune'"))
print(df.query("Age > 30 and ExperienceInCurrentDomain > 3"))

print(df.nlargest(5, "Age")) # using nlargest()
print(df.nsmallest(5, "Age")) # using nsmallest()
```
